This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix');
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed'])
    plt.show()
```

## TASK  3

## TASK  8


Create a decision tree classifier object then  create a  <code>GridSearchCV</code> object  <code>tree_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
parameters = {'criterion': ['gini', 'entropy'],
     'splitter': ['best', 'random'],
     'max_depth': [2*n for n in range(1,10)],
     'max_features': ['auto', 'sqrt'],
     'min_samples_leaf': [1, 2, 4],
     'min_samples_split': [2, 5, 10]}

tree = DecisionTreeClassifier()
```

Create a k nearest neighbors object then  create a  <code>GridSearchCV</code> object  <code>knn_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.

## Calculate Performance Metrics




```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Dictionary of our trained models
models = {
    'Logistic Regression': logreg_cv,
    'SVM': svm_cv,
    'Decision Tree': tree_cv,
    'KNN': knn_cv
}

# List to store metrics for each model
performance_metrics = []

for name, model in models.items():
    # Generate predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    metrics = {
        'Model': name,
        'Accuracy': accuracy_score(Y_test, y_pred),
        'Precision': precision_score(Y_test, y_pred),
        'Recall': recall_score(Y_test, y_pred),
        'F1 Score': f1_score(Y_test, y_pred)
    }
    performance_metrics.append(metrics)

# Display the results
for result in performance_metrics:
    print(result)
```

## Create Summary DataFrame


```python
import pandas as pd

# Create the DataFrame from the performance_metrics list
metrics_df = pd.DataFrame(performance_metrics)

# Set 'Model' as the index
metrics_df.set_index('Model', inplace=True)

# Display the DataFrame to verify structure
print(metrics_df)
```

## Visualize with Grouped Bar Chart



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Reset the index of metrics_df to make 'Model' a regular column
metrics_plot_df = metrics_df.reset_index()

# 2. Use pd.melt to transform the DataFrame into long format
metrics_melted = pd.melt(metrics_plot_df, id_vars='Model', var_name='Metric', value_name='Score')

# 4. Initialize a figure with a suitable size
plt.figure(figsize=(12, 7))

# 5. Use sns.barplot to create a grouped bar chart
sns.set_style("whitegrid")
ax = sns.barplot(data=metrics_melted, x='Metric', y='Score', hue='Model')

# 6. Add title, labels, and legend
plt.title('Falcon 9 Landing Prediction: Model Performance Comparison', fontsize=16)
plt.xlabel('Performance Metrics', fontsize=12)
plt.ylabel('Score (0.0 - 1.0)', fontsize=12)
plt.legend(title='Classification Models', bbox_to_anchor=(1.05, 1), loc='upper left')

# 7. Set y-axis limit and display
plt.ylim(0, 1.0)
plt.tight_layout()
plt.show()
```

## Summary:

### Q&A

**Which model achieved the highest performance across all metrics?**
Based on the latest results, there is no single 'best' model as all four (Logistic Regression, SVM, Decision Tree, and KNN) achieved an identical **Accuracy of 0.833** on the test set. However, the **Decision Tree** achieved the highest **Precision (0.909)**, while the other three models achieved a perfect **Recall (1.0)**.

### Data Analysis Key Findings

*   **Uniform Accuracy:** All tested models reached a plateau with a test accuracy of approximately **83.3%**, suggesting that for this specific test split, the models are converging on similar predictive limits.
*   **Precision vs. Recall Trade-off:**
    *   **Logistic Regression, SVM, and KNN** achieved a perfect **Recall of 1.0**, meaning they correctly identified all successful landings, but at the cost of lower precision (0.8).
    *   **Decision Tree** demonstrated higher **Precision (0.909)**, meaning it was more reliable when it predicted a successful landing, but it had lower recall (0.833).
*   **F1 Score:** Logistic Regression, SVM, and KNN shared a slightly higher **F1 Score of 0.889** compared to the Decision Tree's 0.870, making them slightly better for overall balance between precision and recall.

### Insights or Next Steps

*   **Model Selection:** Since the linear (Logistic), kernel (SVM), and instance-based (KNN) models produced identical results, the choice between them may depend on computational efficiency or interpretability.
*   **Sample Size Limitation:** With only 18 test samples, the metrics are highly sensitive to individual data points. Future work should involve cross-validation on the entire dataset or gathering more data to better differentiate these models.
*   **Feature Engineering:** Given the tie in accuracy, the next step to improve performance would likely be engineering new features rather than further tuning the current algorithms.
